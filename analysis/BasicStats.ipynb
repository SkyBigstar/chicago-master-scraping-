```python
import pandas as pd
```

```python
a_df = pd.read_csv('accidents.csv')
a_df
```

```python
a_df['scraped_at_date'] = a_df['scraped_at'].apply(lambda scraped_at: scraped_at.split(' ')[0])
a_df
```

```python
accident_count_df = pd.DataFrame(a_df['scraped_at_date'].value_counts())
accident_count_df
```

```python
accident_count_df['count'].mean()
```

```python
accident_count_df['count'].median()
```

```python
accident_count_df['count'].std()
```
